```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage import feature
from skimage.metrics import peak_signal_noise_ratio as psnr
from scipy.ndimage import gaussian_laplace
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, Add, Dropout
from tensorflow.keras.utils import Sequence
import tensorflow as tf
import tensorflow_datasets as tfds

# Dataset loader with resizing and normalization
def load_div2k(img_size=(128, 128), num_samples=None):
    dataset = tfds.load("div2k", split="train", as_supervised=True)
    images = []
    for img, _ in dataset.take(num_samples or -1):  # Take specified number of samples or all
        img_resized = tf.image.resize(img, img_size) / 255.0  # Normalize to [0, 1]
        images.append(img_resized.numpy())
    images = np.array(images, dtype=np.float32)
    print(f"Loaded {len(images)} images with shape: {images.shape}")
    return images

# Add random Gaussian noise
def add_noise(images, noise_std=None):
    if noise_std is None:
        noise_std = np.random.uniform(0.05, 0.1)  # Adjusted to match normalized [0, 1] range
    noisy_images = images + noise_std * np.random.randn(*images.shape).astype(np.float32)
    return np.clip(noisy_images, 0, 1)

# Data generator for model training
class DataGenerator(Sequence):
    def __init__(self, clean_images, noisy_images, batch_size=32):
        self.clean_images = clean_images
        self.noisy_images = noisy_images
        self.batch_size = batch_size

    def __len__(self):
        return int(np.ceil(len(self.clean_images) / self.batch_size))

    def __getitem__(self, idx):
        batch_clean = self.clean_images[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_noisy = self.noisy_images[idx * self.batch_size:(idx + 1) * self.batch_size]
        return batch_noisy, batch_clean

# Residual block for DnCNN
def residual_block(input_tensor, filters, kernel_size=3):
    x = Conv2D(filters, kernel_size, padding="same", kernel_initializer="he_normal")(input_tensor)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = Conv2D(filters, kernel_size, padding="same", kernel_initializer="he_normal")(x)
    x = BatchNormalization()(x)
    return Add()([input_tensor, x])

# Build the DnCNN model with improvements
def build_modular_dncnn(depth=25, filters=64, kernel_size=3, dropout_rate=0.1):
    input_img = Input(shape=(None, None, 3))  # RGB input
    x = Conv2D(filters, kernel_size, padding="same", kernel_initializer="he_normal")(input_img)
    x = Activation("relu")(x)
    for _ in range(depth - 2):
        x = residual_block(x, filters, kernel_size)
        x = Dropout(dropout_rate)(x)  # Add dropout for better generalization
    x = Conv2D(3, kernel_size, padding="same", kernel_initializer="he_normal")(x)  # Output RGB
    x = Activation("sigmoid")(x)
    output_img = Add()([input_img, x])  # Residual learning
    model = Model(inputs=input_img, outputs=output_img)
    model.compile(optimizer="adam", loss="mean_squared_error", metrics=["mse"])
    return model

# Apply Canny edge detection
def apply_canny_edge_detection(image):
    edges = feature.canny(image[..., 0], sigma=2)  # Canny on the first channel (R)
    return edges.astype(np.float32)  # Convert to float32 for plotting

# Simulate NSST visualization with improved scaling
def apply_nsst(image):
    if image.dtype != np.float32:
        image = image.astype(np.float32)  # Convert to float32 for compatibility
    nsst_image = gaussian_laplace(image, sigma=3)
    nsst_image = np.abs(nsst_image)  # Use absolute values for better visualization
    nsst_image = (nsst_image - np.min(nsst_image)) / (np.max(nsst_image) - np.min(nsst_image))  # Normalize to [0, 1]
    return nsst_image

# Plot training metrics
def plot_training_metrics(train_loss, val_loss, validation_psnr):
    epochs = range(len(validation_psnr))

    # Plot training and validation loss
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_loss, label='Training Loss')
    plt.plot(epochs, val_loss, label='Validation Loss')
    plt.title('Loss Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot validation PSNR
    plt.subplot(1, 2, 2)
    plt.plot(epochs, validation_psnr, label='Validation PSNR', color='orange')
    plt.title('PSNR Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('PSNR')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Main pipeline
def main_pipeline():
    # Step 1: Load dataset
    print("Loading dataset...")
    train_images = load_div2k(img_size=(128, 128), num_samples=1000)
    test_images = load_div2k(img_size=(128, 128), num_samples=200)

    # Step 2: Add noise
    print("Adding noise to images...")
    noisy_train_images = add_noise(train_images)
    noisy_test_images = add_noise(test_images)

    # Step 3: Create data generators
    print("Creating data generators...")
    train_generator = DataGenerator(train_images, noisy_train_images, batch_size=8)

    # Step 4: Build DnCNN model
    print("Building DnCNN model...")
    modular_dncnn = build_modular_dncnn()

    # Step 5: Train the model and calculate PSNR
    print("Training the model...")
    train_loss, val_loss, validation_psnr = [], [], []
    for epoch in range(50):
        history = modular_dncnn.fit(
            train_generator,
            epochs=1,  # Train one epoch at a time to track PSNR
            validation_data=(noisy_test_images, test_images),
            verbose=1
        )
        train_loss.append(history.history['loss'][0])
        val_loss.append(history.history['val_loss'][0])

        # Calculate PSNR for validation images
        denoised_test_images = modular_dncnn.predict(noisy_test_images)
        psnr_epoch = np.mean([psnr(test_images[i], denoised_test_images[i]) for i in range(len(test_images))])
        validation_psnr.append(psnr_epoch)
        print(f"Epoch {epoch + 1}: Validation PSNR = {psnr_epoch:.2f}")

    # Step 6: Plot training metrics
    plot_training_metrics(train_loss, val_loss, validation_psnr)

    # Step 7: Evaluate and visualize
    print("Evaluating the model...")
    noisy_test_image = noisy_test_images[0]  # Use the first test image
    clean_test_image = test_images[0]  # Corresponding clean image
    denoised_test_image = modular_dncnn.predict(noisy_test_image[np.newaxis, ...])[0]

    # Plot results
    plt.figure(figsize=(15, 5))
    titles = ["Noisy Image", "Clean Image", "Denoised Image", "Canny Edges", "NSST"]
    images = [
        noisy_test_image,
        clean_test_image,
        denoised_test_image,
        apply_canny_edge_detection(denoised_test_image),
        apply_nsst(denoised_test_image),
    ]

    for i, (title, img) in enumerate(zip(titles, images)):
        plt.subplot(1, 5, i + 1)
        plt.title(title)
        if i < 3:  # First 3 images are RGB
            plt.imshow(img)
        else:  # Canny and NSST images are grayscale
            plt.imshow(img, cmap="gray")
        plt.axis("off")

    plt.tight_layout()
    plt.show()

# Run the pipeline
if __name__ == "__main__":
    main_pipeline()
```
